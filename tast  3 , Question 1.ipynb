```python
import numpy as np
from IPython.display import display
import pandas as pd
```

```python
df_Items = pd.read_csv("Items.csv")
display(df_Items)

#   
# - 
display(df_Items.info())

# - 
display(df_Items.describe())

# -
display(df_Items.head(5))


#   

# -
def count_null(row):
    count = 0
    for col in df_Items.columns:
        if pd.isna(row[col]):
            count += 1
    return count
df_Items['nan_count'] = df_Items.apply(count_null, axis=1)
display(df_Items)

# -
df_Items.sort_values(by=['item_name', 'nan_count'], ascending=[True, True])

# -
display(df_Items['id'].drop_duplicates())
```
